TRAIN VAL TEST SPLIT

```python
from sklearn.model_selection import train_test_split
```

```python
train_val_df, test_df = train_test_split(raw_df, test_size=0.2, random_state=1)
train_df, val_df = train_test_split(train_val_df, test_size=0.25, random_state=1)
```

```python
print("train_df.shape: ", train_df.shape)
print("val_df.shape: ", val_df.shape)
print("test_df.shape: ", test_df.shape)
```

```python
input_cols = ['animal_name', 'species', 'hair', 'feathers', 'eggs',
       'milk', 'airborne', 'aquatic', 'predator', 'toothed', 'backbone',
       'breathes', 'venomous', 'fins', 'legs', 'tail', 'domestic', 'catsize']
target_col = 'class_type'
```

```python
print(input_cols)
```

```python
print(target_col)
```

```python
train_inputs = train_df[input_cols].copy()
train_targets = train_df[target_col].copy()
```

```python
train_targets
```

```python
val_inputs = val_df[input_cols].copy()
val_targets = val_df[target_col].copy()
```

```python
test_inputs = test_df[input_cols].copy()
test_targets = test_df[target_col].copy()
```

```python
print(val_inputs.shape)
print(val_targets.shape)
```

```python
print(test_inputs.shape)
print(test_targets.shape)
```

```python
import numpy as np
```

IDENTIFYING THE COLUMNS WITH NUMERIC AND TEXT VALUES

```python
numeric_cols = train_inputs.select_dtypes(include=np.number).columns.tolist()
categorical_cols = train_inputs.select_dtypes('object').columns.tolist()
```

ENCODING THE TEXT COLUMNS USING THE ONE HOT ENCODER

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
```

```python
encoder.fit(raw_df[categorical_cols])
```

```python
encoder.categories_
```

```python
raw_df[categorical_cols]
```

```python
encoded_cols = list(encoder.get_feature_names(categorical_cols))
print(encoded_cols)
```

USING THE ENCODER ON THE TRAIN, VAL AND TEST INPUT DATAFRAMES

```python
train_inputs[encoded_cols] = encoder.transform(train_inputs[categorical_cols])
val_inputs[encoded_cols] = encoder.transform(val_inputs[categorical_cols])
test_inputs[encoded_cols] = encoder.transform(test_inputs[categorical_cols])
```

DROPPING THE COLUMNS WITH TEXT DATA AND CREATING NEW TRAIN, VAL AND TEST DATAFRAMES AS: X_train, X_val and X_test

IMPORTING THE XGBOOST, INITIALIZING THE XGBOOST CLASSIFIER AND FITTING THE MODEL

```python
import xgboost
```

```python
model = xgboost.XGBClassifier()
model.fit(X_train, train_targets)
```

MAKING PREDICTIONS ON THE FITTED MODEL
